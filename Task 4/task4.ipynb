### Visualization of the data set

```python
sns.pairplot(df,hue='Category')
plt.show()
```

```python

ax = sns.countplot(df['Ship Mode'],data=df)
plt.title('Ship Mode',fontsize=30)
```

```python
ax = sns.countplot(df['Segment'],data=df)
plt.title('Segment',fontsize=30)
```

```python
ax = sns.countplot(df['Region'],data=df)
plt.title('Region',fontsize=30)
```

```python
df["Profit"].plot();
```

```python
df.hist(figsize=(12,17),bins=30);
```

```python
df.plot.box() 
sns.set(rc={'figure.figsize':(5,11)})
plt.title("Visualizing the annomalities,Mean & Median",fontsize=20)

# individual attribute box plot 
plt.boxplot(df['Sales']) 
plt.show() 
```

```python

sns.set(rc={'figure.figsize':(10,10)})
plt.title("Scatter plot Between Quantity & Sales ",fontsize=30)

plt.scatter(df['Quantity'],df['Sales']) 
plt.show()
```

```python
sns.set(rc={'figure.figsize':(10,10)})
plt.title("Scatter plot Between Discount & Sales ",fontsize=30)


plt.scatter(df['Discount'],df['Sales']) 
plt.show()
```

```python

sns.heatmap(df.corr(), annot=True);
```

```python
df.corr().plot.bar();
```

```python
sns.countplot(x="Category", data=df).set_title("Most sold product line");
```

```python
sns.countplot(x="Sub-Category", data=df).set_title("Most sold Sub-Category");
plt.xticks(rotation=90);
```

```python

# Let's plot a relationship between sales and profit
plt.scatter(x=df['Sales'], y=df['Profit'])
plt.xlabel("Total Sales")
plt.ylabel("Profit")
plt.show()
```

```python
sns.countplot(x='Region',hue='Category',data=df)
```
